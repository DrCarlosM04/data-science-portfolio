```python
import pandas as pd

df = pd.read_csv("nanotox_dataset.csv")
df.head()
```

```python
df.info()
df.describe()
```

NPs column indicates the type of nanoparticle whose data we have in the table. For each type and variation of nanoparticle we know their coresize: size of the particle in nm, hydrosize: hydrodynamic size of the particle (probably measured by DLS); surfcharge: charge on the surface of the particle (probably measured with a potential Z-meter); surfarea: the surface area of the nanoparticles; Ec: electric charge on the particle; Expotime: Exposure time; dosage: the ammount of material used; e: energy related feature; NOxygen: number of oxygen atoms; class: label indicating wether the nanoparticles are toxic or non-toxic.

The units for each variable are coresize: nanometers; hydrosize: nm; surfcharge: mV; Ec: microCoulombs; Expotime: seconds; dosage: mg/k

## Types of nanoparticles and average properties 

```python
df.groupby(["NPs", "class"])[
    "coresize", "surfcharge", "Ec", "surfarea", "NOxygen"
].mean()
```

There are some particles that are never found to be toxic, like Fe2O3 and Al2O3. The other three types can be both toxic or non-toxic. This will depend on their physical and chemical properties

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create dispersion matrix
sns.pairplot(df, hue="NPs", kind="reg")
plt.show()
```

The majority of data comes from TiO2 and ZnO particles. The rest of the observations are not enough to draw conclussions. The next steps would be to analyze the differences between Toxic and Nontoxic particles to try and see if any property correlates with the toxicity and which properties do not affect the toxicity of the materials.
I would not consider Fe2O3 or Al2O3 since there are not information of those materials being toxic.

```python
df_CuO = df[df["NPs"] == "CuO"]
df_TiO2 = df[df["NPs"] == "TiO2"]
df_ZnO = df[df["NPs"] == "ZnO"]


for d in (df_CuO, df_TiO2, df_ZnO):
    sns.pairplot(d, hue="class", kind="reg")
    plt.show()
```

Even when analyzing each type of particle independently and classifying them as toxic or nontoxic there are not clear visible trends other than expected such as the physical properties being linearly correlated. However, there is no enough evidence to visually catalogue as toxic or nontoxic any of the particles only from their physical properties.
